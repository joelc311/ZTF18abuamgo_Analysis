```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from astropy.io import fits

directory = '/Users/joel/anaconda3/lib/python3.11/site-packages/pyqsofit/pca/Yip_pca_templates/'
templates = ['gal_eigenspec_Yip2004.fits',
             'qso_eigenspec_Yip2004_AZBIN4.fits',
             'qso_eigenspec_Yip2004_AZBIN5.fits',
             'qso_eigenspec_Yip2004_BZBIN2.fits',
             'qso_eigenspec_Yip2004_BZBIN3.fits',
             'qso_eigenspec_Yip2004_BZBIN4.fits',
             'qso_eigenspec_Yip2004_BZBIN5.fits',
             'qso_eigenspec_Yip2004_CZBIN1.fits',
             'qso_eigenspec_Yip2004_CZBIN2.fits',
             'qso_eigenspec_Yip2004_CZBIN3.fits',
             'qso_eigenspec_Yip2004_CZBIN4.fits',
             'qso_eigenspec_Yip2004_DZBIN1.fits',
             'qso_eigenspec_Yip2004_DZBIN2.fits',
             'qso_eigenspec_Yip2004_global.fits']
```

```python
for temp in templates:
    hdul = fits.open(directory+temp)
    

    x = hdul[1].data[0][0]
    y = hdul[1].data[0][1][0]
    
    hdul.close()
    
    plt.plot(x, y)
    plt.title(temp)
    plt.show()
```
